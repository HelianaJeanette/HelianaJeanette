Importar librerias

```python
import requests
import json 
```

Realizar llamada a la API

```python
url='https://datos.gob.es/apidata/catalog/dataset'

r = requests.get(url+"?"+"_sort=title&_pageSize=10&_page=0")
```

Imprimir cabecera

```python
print("*****************Cabeceras**************+")
print(r.headers)
```

Imprimir valores

```python
json_object = r.json()

print("*******************Datos***************")
print (json.dumps(json_object, indent=4, sort_keys=True))
```

Imprimir el valor 4



```python
print(json_object[4])
```
